```python
# generate csv files for train and test files
def generate_csv(files,csv_name):
    df_text=pd.DataFrame(columns=['content','label'])
    for i,file in enumerate(files):
        label=file.split('\\')[-2]

        with open(file) as f:
            data = f.read()
            df_text.loc[i]={'content':data,'label':label}
    df_text.to_csv('data\\'+csv_name+'.csv')
```

## Data cleaning

1. remove punctuations, numbers and other non-words.
2. lemmatize words
3. convert to lower case
4. select words in GloVe

```python
import re
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer 

def pre_process(df_text):
    wnl = WordNetLemmatizer()

    df_text.content =df_text.content.apply(lambda x: re.sub('\W+',' ', x))
    df_text.content =df_text.content.apply(lambda x: wnl.lemmatize(x))
    df_text.content =df_text.content.apply(lambda x: x.lower())
    return df_text
```

```python
df_train=pre_process(df_train)
df_test=pre_process(df_test)
df_train.head()
```

Since I want to use GloVe in some models, I have to remove the words not in GloVe dictionary.

```python
def GloVe_dict(GloVe_filename):
    word_vector={}
    with open(GloVe_filename, encoding="utf-8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            word_vector[word] = coefs
    return word_vector

GloVe_filename='data\\glove.6B.100d.txt'
word_vectors=GloVe_dict(GloVe_filename)
```

```python
# remove words not in GloVe
df_test.content =df_test.content.apply(lambda x: ' '.join([s for s in x.split(' ') if s in word_vectors.keys()]))
df_train.content =df_train.content.apply(lambda x: ' '.join([s for s in x.split(' ') if s in word_vectors.keys()]))
df_train.head()
```

Most texts have less than 2000 words, but some texts have too many words. I can cut one text into many texts to augment data. If the text is too short, I'll duplicate the text in order to ensure the same weight of each text.

```python
# cut one text into many texts
def cut_text(df_text,num,max_len):
    df_tot=pd.DataFrame(columns=df_text.columns)
    k=0
    for i,content in enumerate(df_text.content):
        word_list=content.split(' ')
        content_len=len(word_list)
        # if the length of content is larger than required length 
        if content_len>max_len:
            # find the last cut location
            end_loc=content_len-max_len
            # use equal intervals between 0 and last cut location to find start locations  
            start_locs=np.linspace(0, end_loc, num)
            for loc in start_locs:
                # a piece of text with required length, some texts may cross over
                new_content=' '.join(word_list[int(loc):int(loc)+max_len])
                df_tot.loc[k]={'content':new_content,'label':df_text.label.loc[i]}
                k+=1
        else:
            # duplicate short texts to make sure all the texts have the same weight
            for j in range(num):
                df_tot.loc[k]={'content':content,'label':df_text.label.loc[i]}
                k+=1
    return df_tot
```

```python
# each text is divided into 5 texts, each contains at most 300 words
df_train_processed=cut_text(df_train,5,300)
df_test_processed=cut_text(df_test,5,300)
df_train_processed.to_csv('data/train_processed.csv')
df_test_processed.to_csv('data/test_processed.csv')
```

#  Naive Bayes

- I firstly use Naive Bayes model as a baseline.

## Split dataset

```python
# split train set into train and validation
from sklearn.model_selection import train_test_split

def split_train(df,test_size):
    x = df['content'].values
    y = df['label'].values

    x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=test_size,random_state=42)
    print('Size of train, validation:', len(y_train), len(y_val))
    return x_train, x_val, y_train, y_val
```

```python
# use train_clean.csv since the model doesn't require the texts to have same length. 
df_train= pd.read_csv('data\\train_clean.csv',index_col=0)
x_train, x_val, y_train, y_val=split_train(df_train,0.2)
```

## Build model 

```python
# use CountVectorizer and TfidfVectorizer to convert words into numbers
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

def naive_bayes(x_train,y_train,x_val,alpha=1.0):
    vect = TfidfVectorizer(max_features=2000000, stop_words='english', min_df=0.002, max_df=0.5, 
                            ngram_range=(1,3),tokenizer=lambda x: x.split(' '))
    vect.fit(x_train)
    X_dtm = vect.transform(x_train)
    X_dtm_val = vect.transform(x_val)

    clf = MultinomialNB(alpha=alpha)
    clf.fit(X_dtm, y_train)
    y_val_pred = clf.predict(X_dtm_val)
    y_train_pred = clf.predict(X_dtm)
    return y_train_pred, y_val_pred
```

```python
y_train_pred, y_val_pred=naive_bayes(x_train,y_train,x_val,alpha=1.0)
```

## Performance 

```python
from sklearn.metrics import confusion_matrix,accuracy_score,f1_score
import matplotlib.pyplot as plt

# show the confusion matrix, accuracy score and macro-average f1 score 
def performance(y_pred,y_true):
    
    confusion = confusion_matrix(y_true,y_pred)
    plt.figure(figsize=(30,10))
    plt.imshow(confusion, cmap=plt.cm.Blues)
    indices = range(len(confusion))
    plt.colorbar()
    plt.xlabel('guess')
    plt.ylabel('fact')
    for first_index in range(len(confusion)):
        for second_index in range(len(confusion[first_index])):
            plt.text(first_index, second_index, confusion[first_index][second_index])

    plt.show()
    fscore=f1_score(y_true, y_pred, average='macro', sample_weight=None)
    print("Accuracy score is:",accuracy_score(y_true,y_pred))
    print("The macro-average of all classes f1 score is:",fscore)
```

```python
performance(y_train_pred,y_train)
performance(y_val_pred,y_val)
```

## Tune hyper parameters

I only tune one parameter 'alpha' in MultinomialNB model.

```python
# tune the parameter alpha according to validation set and return the best parameter
def tune_nb(x_train,y_train,x_val,y_val,alphas):
    fscores=[]
    for alpha in alphas:
        y_train_pred, y_val_pred=naive_bayes(x_train,y_train,x_val,alpha=alpha)
        fscore=f1_score(y_val, y_val_pred, average='macro', sample_weight=None)
        fscores.append(fscore)

    plt.plot(alphas,fscores,'.r-')
    plt.legend()
    plt.xlabel('alpha')
    plt.ylabel('f1_score')
    return alphas[fscores.index(max(fscores))]
```

```python
alphas=[0,0.1,0.5,1,5]
alpha_best=tune_nb(x_train,y_train,x_val,y_val,alphas)
```

## Tokenizing and Padding

Tokenize the texts and pad each text into the same length  

```python
# convert dataframe to list of texts and labels
def extract_text(df):
    texts = []
    labels_data = []
    for i in range(df['content'].shape[0]):
        text = df['content'][i]
        texts.append(text)
        labels_data.append(df.label[i])
    return texts,labels_data
```

```python
# tokenize words into index for train set and test set 
def tokenize_text(train_text,test_text):
    # Tokenizer
    tokenizer = Tokenizer(num_words=MAX_NB_WORDS, oov_token='<UNK>')
    tokenizer.fit_on_texts(train_text)
    # list of lists with different length
    train_idxs = tokenizer.texts_to_sequences(train_text)
    test_idxs = tokenizer.texts_to_sequences(test_text)   

    word_index = tokenizer.word_index  # word to index, such as {'the': 1, 'a':2}
    
    return train_idxs,test_idxs,word_index
```

```python
# pad each sequence into same length
def pad_text(data_idxs,labels_data):
    from sklearn.preprocessing import LabelEncoder
    # transform 20 string labels into 20 numbers
    label_array = np.array(labels_data)
    label_encoder = LabelEncoder()
    label_int = label_encoder.fit_transform(label_array) 

    word_index_matrix = pad_sequences(data_idxs, maxlen=MAX_SEQUENCE_LENGTH)   # lose some words
    labels_matrix = to_categorical(label_int)
    print('Shape of word index matrix:', word_index_matrix.shape)
    print('Length of label matrix:', labels_matrix.shape)
    return word_index_matrix, labels_matrix
```

```python
df_train=pd.read_csv('data\\train_clean.csv',index_col=0)
df_test=pd.read_csv('data\\test_clean.csv',index_col=0)

train_text,train_labels_data=extract_text(df_train)
test_text,test_labels_data=extract_text(df_test)

train_idxs,test_idxs,word2index=tokenize_text(train_text,test_text)

word_index_train, labels_train=pad_text(train_idxs,train_labels_data)
x_test, y_test=pad_text(test_idxs,test_labels_data)
```

```python
def bert_embed(word_index):
    from bert_embedding import BertEmbedding
    
    bert_embedding = BertEmbedding()
    embedding_matrix = np.random.random((len(word_index) + 1, EMBEDDING_DIM))
    for word, i in word_index.items():
        # record the process
        if i%100==0:
            print(i)
        result=bert_embedding([word])
        embedding_vector = result[0][1][0]
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
    return embedding_matrix
```

```python
def generate_embedding(GloVe_filename,word_index,useBert=False):
    #Use Pre-trained GloVe word embeddings or Bert embedding
    embeddings_index = {}
    embedding_matrix = np.random.random((len(word_index) + 1, EMBEDDING_DIM)) 

    if useBert==True:
        embedding_matrix= bert_embed(word_index)

    else:
        GloVe_file = open(GloVe_filename, encoding="utf-8")
        for line in GloVe_file:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
        f.close()
        print('Total %s word vectors in Glove.' % len(embeddings_index))
    
        # assign a vector to a word if it's in GloVe  
        for word, i in word_index.items():
            embedding_vector = embeddings_index.get(word)
            if embedding_vector is not None:
                # words not found in embedding index will be random.
                embedding_matrix[i] = embedding_vector

    # embedding layer converts each word into a vector
    embedding_layer = Embedding(len(word_index) + 1,
                                EMBEDDING_DIM,
                                weights=[embedding_matrix],
                                input_length=MAX_SEQUENCE_LENGTH)
    return embedding_layer
```

## Build the network 

I build a neural network with embedding, Conv, LSTM and Dense layers.

```python
embedding_layer.trainable=False  # use GloVe vectors, don't train this layer 

lstm_model = Sequential()
lstm_model.add(embedding_layer)
lstm_model.add(Dropout(0.2))
lstm_model.add(Conv1D(filters=32, kernel_size=5, padding='same', activation='relu'))
lstm_model.add(MaxPooling1D(pool_size=2))
lstm_model.add(Conv1D(filters=64, kernel_size=3, padding='same', activation='relu'))
lstm_model.add(MaxPooling1D(pool_size=2))
lstm_model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
lstm_model.add(BatchNormalization())
lstm_model.add(Dense(256, activation='relu'))
lstm_model.add(Dense(128, activation='relu'))
lstm_model.add(Dense(64, activation='relu'))
lstm_model.add(Dense(20, activation='softmax'))
lstm_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(lstm_model.summary())
```
