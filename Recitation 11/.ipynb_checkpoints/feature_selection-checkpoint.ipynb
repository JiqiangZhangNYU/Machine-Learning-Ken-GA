# Feature Selection Methods 

Filter：score all features according to correlation and distribution, select features according to thresholds.

Wrapper：according to target function results, select or remove features

Embedded：train some ML models to determine the weight of each feature, then select features

```python
from sklearn.datasets import load_iris
 
iris = load_iris()

iris
```

```python
X=iris.data
y=iris.target
```

# Filter 

## variance selection

```python
from sklearn.feature_selection import VarianceThreshold

# threshold for variance
VarianceThreshold(threshold=0.5).fit_transform(iris.data)
```

## Correlation 

```python
from scipy.stats import pearsonr

X=iris.data
y=iris.target
for i in range(X.shape[1]):
    print(pearsonr(X[:,i],y))
```

## chi square 

Recall that the chi-square test measures dependence between stochastic variables, so using this function “weeds out” the features that are the most likely to be independent of class and therefore irrelevant for classification.

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

SelectKBest(chi2, k=2).fit_transform(iris.data, iris.target)
```

# Wrapper 

## recursive feature elimination 

Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), the goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through a coef_ attribute or through a feature_importances_ attribute. Then, the least important features are pruned from current set of features. That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

RFE(estimator=LogisticRegression(), n_features_to_select=2).fit_transform(iris.data, iris.target)
```

# Embedded 

## Linear Model l1

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression

# L1 penalty for feature selection 
SelectFromModel(LogisticRegression(penalty="l1", C=0.1)).fit_transform(iris.data, iris.target)
```

##  Tree model

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import GradientBoostingClassifier

#GBDT feature importance
SelectFromModel(GradientBoostingClassifier()).fit_transform(iris.data, iris.target)
```

```python
gbc=GradientBoostingClassifier()
gbc.fit(X,y)
gbc.feature_importances_
```

```python
from skfeature.function.statistical_based import CFS
from sklearn import svm
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import SelectKBest
from sklearn.pipeline import Pipeline

def test_cfs():
    # load data
    from sklearn.datasets import make_classification
    X, y = make_classification(n_samples=200, n_features=20, n_informative=5, n_redundant=5, n_classes=2)
    X = X.astype(float)
    n_samples, n_features = X.shape    # number of samples and number of features

    num_fea = 3

    # build pipeline
    pipeline = []
    pipeline.append(('select top k', SelectKBest(score_func=CFS.cfs, k=num_fea)))
    pipeline.append(('linear svm', svm.LinearSVC()))
    model = Pipeline(pipeline)
    
    # split data into 5 folds
    kfold = KFold(n_splits=5, shuffle=True)
    results = cross_val_score(model, X, y, cv=kfold)
    print("Accuracy: {}".format(results.mean()))

test_cfs()
```

```python
from skfeature.function.information_theoretical_based import DISR
from sklearn import svm
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import SelectKBest
from sklearn.pipeline import Pipeline

def test_disr():
    from sklearn.datasets import make_classification
    X, y = make_classification(n_samples=200, n_features=20, n_informative=5, n_redundant=5, n_classes=2)
    X = X.astype(float)
    n_samples, n_features = X.shape    # number of samples and number of features

    num_fea = 3

    # build pipeline
    pipeline = []
    pipeline.append(('select top k', SelectKBest(score_func=DISR.disr, k=num_fea)))
    pipeline.append(('linear svm', svm.LinearSVC()))
    model = Pipeline(pipeline)

    # split data into 5 folds
    kfold = KFold(n_splits=5, shuffle=True)
    results = cross_val_score(model, X, y, cv=kfold)
    print("Accuracy: {}".format(results.mean()))

test_disr()
```
